## Importar las librerias al notebook

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Hacer que leea o que guarde nuestros datos proporcionados

```python
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/mpg.csv"
df = pd.read_csv(url)
```


Imprime las primeras 5 filas y usa un comando para mostrar los tipos de datos de cada columna.

```python
#imprimimos los primeros 5 datos
print ("Primeros 5 datos proporcionados de la lista")
print(df.head())
```

Identifica si hay valores nulos en la columna horsepower (caballos de fuerza).

```python
print("Mostrar o identificar los datos de las columnlas")
print(df.info())
```

Reemplaza esos valores nulos con la mediana de esa misma columna usando NumPy y Pandas.

```python
valoresnulos = df['horsepower'].isnull().sum()
print(f"\nValores nulos de los datos:{valoresnulos}")
```

Crea una nueva columna llamada relacion_peso_potencia que sea el resultado de dividir la columna weight (peso) entre horsepower.

```python
mediana_hp=np.nanmedian(df['horsepower'])
```

```python
df['horsepower']=df['horsepower'].fillna(mediana_hp)
```

```python
print(f"Valoresnulos que se reemplazaron:{mediana_hp}")
```

```python
df['relacion_peso_potencia'] = df['weight'] / df['horsepower']
```

```python
print("\nDataFrame con la nueva columna 'relacion_peso_potencia':")
print(df[['name', 'weight', 'horsepower', 'relacion_peso_potencia']].head())
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df['acceleration'], bins=20, color='green')
plt.title('Distribucion de la aceleracion')
plt.xlabel('aceleracion')
plt.ylabel('frecuencia o tiempo')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(df['displacement'], df['horsepower'], color='skyblue')
plt.title('desplazamiento del motor y Hp')
plt.xlabel('desplazamiento')
plt.ylabel('caballos de Fuerza (HP)')
plt.grid(True)
plt.show()
```

```python
promedio_mpg_por_pais = df.groupby('origin')['mpg'].mean()
```

```python
plt.figure(figsize=(8, 5))
promedio_mpg_por_pais.plot(kind='bar', color=['skyblue', 'lightgreen', 'green'])
plt.title('promedio de Rendimiento por pais de Origen')
plt.xlabel('pais de origen')
plt.ylabel('Millas por Galón (MPG) Promedio')
plt.show()
```
